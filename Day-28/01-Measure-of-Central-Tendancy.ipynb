```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Generate random number from a skewed normal distribution

x = np.random.standard_t(10, size = 1000)
x[:15]
```

```python
plt.hist(x, color = 'red')
```

```python
mean = np.mean(x)
print("Mean: ", mean)
```

```python
plt.hist(x, color = 'red')
plt.axvline(x = mean, color= 'black')
```

```python
median = np.median(x)
print("Median: ", median)
```

```python
plt.hist(x, color = 'red')
plt.axvline(x = median, color= 'black')
```

### Mode

```python
mode = stats.mode(x).mode
print("Mode:", mode)
```

```python
plt.hist(x,color = 'red')
plt.axvline(x = mode, color = 'black')
```
