```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#Generate x values for the standart normal distribution
x=np.linspace(-4,4,1000)

#calculate the critical z-scores
#ppf = percent point function
z_critical = stats.norm.pdf(1-0.05/2)

#plot the standard normal distribution
plt.figure(figsize=(8,6))
plt.plot(x,stats.norm.pdf(x), 'blue', lw=2, label='Standard Normal Distribution')

#Shade the rejection region
plt.fill_between(x, stats.norm.pdf(x), where=(x<=-z_critical) | (x>=z_critical), color = 'red', alpha = 0.3, label='Rejection Region')

#Add labels and title
plt.title('Standard Normal Distribution with Rejection Region')
plt.xlabel('Z-score')
plt.ylabel('Density')
plt.legend()
plt.show()
```
